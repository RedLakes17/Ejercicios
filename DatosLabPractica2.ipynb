```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy.signal import find_peaks

# Cargar datos
df = pd.read_csv('TSAZ01AM.csv', sep='\s+', header=None)
df2 = pd.read_csv('TSAZ01BM.csv', sep='\s+', header=None)

# Función para obtener máximos y mínimos locales
def max_min_local(x, y):
    peaks, _ = find_peaks(y)
    valleys, _ = find_peaks(-y)
    return peaks, valleys

# Obtener máximos y mínimos
peaks1, valleys1 = max_min_local(df[0], df[1])
peaks2, valleys2 = max_min_local(df2[0], df2[1])

# Filtrar para que solo consideren λ ≥ 400 nm
peaks1 = [i for i in peaks1 if df[0][i] >= 400]
valleys1 = [i for i in valleys1 if df[0][i] >= 400]
peaks2 = [i for i in peaks2 if df2[0][i] >= 400]
valleys2 = [i for i in valleys2 if df2[0][i] >= 400]

# Imprimir resultados
print("TSAZ01AM - Máximos locales ≥400nm:", df[0][peaks1].values)
print("TSAZ01AM - Mínimos locales ≥400nm:", df[0][valleys1].values)
print("TSAZ01BM - Máximos locales ≥400nm:", df2[0][peaks2].values)
print("TSAZ01BM - Mínimos locales ≥400nm:", df2[0][valleys2].values)

# Graficar líneas principales
plt.plot(df[0], df[1], linestyle='-', color='green', label='TSAZ01AM')
plt.plot(df2[0], df2[1], linestyle='-', color='orange', label='TSAZ01BM')

# Marcar máximos y mínimos (solo ≥400nm)
plt.plot(df[0][peaks1], df[1][peaks1], "go", label='Máx TSAZ01AM')
plt.plot(df[0][valleys1], df[1][valleys1], "gs", label='Mín TSAZ01AM')
plt.plot(df2[0][peaks2], df2[1][peaks2], "ro", label='Máx TSAZ01BM')
plt.plot(df2[0][valleys2], df2[1][valleys2], "rs", label='Mín TSAZ01BM')

# Etiquetas, título y grilla
plt.xlabel('Longitud de onda (nm)')
plt.ylabel('Transmitancia')
plt.title('Transmitancia con Máximos y Mínimos Locales del TSAZ01')
plt.grid(True)

# Leyenda fuera del gráfico
plt.legend(loc='upper center', bbox_to_anchor=(0.5, -0.15), ncol=2)

plt.show()
```
